# Preparing Data

```python
x=df.drop('Class',axis=1)
y=df['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

## Model Evaluation

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
```

```python
classifier={
    "LogisticRregression": LogisticRegression(),
    "Decision Tree Classifier": DecisionTreeClassifier()
}

for name,clf in classifier.items():
    print(f"\n========{name}=========")
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)
    print(f"\nAccuracy=:{accuracy_score(y_test,y_pred)}")
    print(f"\nPrecision=:{precision_score(y_test,y_pred)}")
    print(f"\nRecall=:{recall_score(y_test,y_pred)}")
    print(f"\nF1 Score=:{f1_score(y_test,y_pred)}")
```

# Over Sampling

```python
x=df.drop('Class',axis=1)
y=df['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_res,y_res=SMOTE().fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_res,y_res,test_size=0.2,random_state=42)
```

```python
classifier={
    "LogisticRregression": LogisticRegression(),
    "Decision Tree Classifier": DecisionTreeClassifier()
}

for name,clf in classifier.items():
    print(f"\n========{name}=========")
    clf.fit(x_train,y_train)
    y_pred=clf.predict(x_test)
    print(f"\nAccuracy=:{accuracy_score(y_test,y_pred)}")
    print(f"\nPrecision=:{precision_score(y_test,y_pred)}")
    print(f"\nRecall=:{recall_score(y_test,y_pred)}")
    print(f"\nF1 Score=:{f1_score(y_test,y_pred)}")
```

# Confusion Matrix

Confusion Matrix Components:
It visually represents the predicted versus actual classifications, showing the counts of True Positives (TP), True Negatives (TN), False Positives (FP), and False Negatives (FN). 




Purpose:
This matrix helps assess how well a classification model distinguishes between different classes (e.g., "Normal" and "Fraud" in this case).

```python
import seaborn as sns
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",xticklabels=['Normal', 'Fraud'], yticklabels=['Normal', 'Fraud'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.show()
```

143 (False Positives): The model incorrectly predicted "Fraud" when the true class was "Normal."

67 (False Negatives): The model incorrectly predicted "Normal" when the true class was "Fraud."

54936 (True Positives): The model correctly predicted "Fraud" when the true class was "Fraud."
